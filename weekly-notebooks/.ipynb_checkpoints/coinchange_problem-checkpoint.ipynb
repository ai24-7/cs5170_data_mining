```python
#include <iostream>
#include <fstream>
#include <vector>
#include <ctime>

int coin_change_combinations_1(std::vector<int> coins, int amount) {
    int num_coins = coins.size();
    
    std::vector<int> ways(amount + 1, 0);
    
    ways[0] = 1;
    
    for (int i = 0; i < num_coins; i++) {
        int coin_value = coins[i];
        
        for (int j = coin_value; j <= amount; j++) {
            ways[j] += ways[j - coin_value];
        }
    }
    
    return ways[amount];
}

void generate_combinations(int index, int remaining_amount, std::vector<int> coins, std::vector<int> current_combination, std::vector<std::vector<int>>& combinations) {
    if (remaining_amount == 0) {
        combinations.push_back(current_combination);
        return;
    }
    
    for (int i = index; i < coins.size(); i++) {
        int coin = coins[i];
        
        if (remaining_amount < coin) {
            continue;
        }
        
        current_combination.push_back(coin);
        
        generate_combinations(i, remaining_amount - coin, coins, current_combination, combinations);
        
        current_combination.pop_back();
    }
}

std::vector<std::vector<int>> coin_change_combinations_2(std::vector<int> coins, int amount) {
    std::vector<std::vector<int>> combinations;
    generate_combinations(0, amount, coins, std::vector<int>(), combinations);
    return combinations;
}

std::vector<double> run_experiment(std::vector<int> coins, std::vector<int> amounts, int (*method)(std::vector<int>, int)) {
    std::vector<double> runtimes;
    for (int amount : amounts) {
        clock_t start_time = clock();
        int res = method(coins, amount);
        double elapsed_time = (clock() - start_time) / (double)CLOCKS_PER_SEC;
        runtimes.push_back(elapsed_time);
    }
    return runtimes;
}


int main() {
    std::vector<int> coins = {1, 5, 10, 25, 50, 100, 200, 500, 1000, 2000};
    std::vector<int> coins_2 = {1, 29, 493};
    std::vector<int> experiment_amounts_2ab = {10, 50, 100, 500, 1000, 1500, 2000, 3000, 5000};
    std::vector<int> experiment_amounts_2c = {10, 25, 50, 100};

    // Run experiments
    std::vector<double> runtimes_2a = run_experiment(coins, experiment_amounts_2ab, coin_change_combinations_1);
    std::vector<double> runtimes_2b = run_experiment(coins_2, experiment_amounts_2ab, coin_change_combinations_1);
    std::vector<double> runtimes_2ca = run_experiment(coins, experiment_amounts_2c, coin_change_combinations_2);
    std::vector<double> runtimes_2cb = run_experiment(coins_2, experiment_amounts_2c, coin_change_combinations_2);

    // Create and open the CSV file
    std::ofstream csvFile("experiment_results.csv");

    // Write header to CSV
    csvFile << "coins,coins_2,experiment_amounts_2ab,experiment_amounts_2c,runtimes_2a,runtimes_2b,runtimes_2ca,runtimes_2cb\n";

    // Write data to CSV
    for (size_t i = 0; i < experiment_amounts_2ab.size(); ++i) {
        csvFile << coins[i] << "," << coins_2[i] << "," << experiment_amounts_2ab[i] << "," << experiment_amounts_2c[i]
                << "," << runtimes_2a[i] << "," << runtimes_2b[i] << "," << runtimes_2ca[i] << "," << runtimes_2cb[i] << "\n";
    }

    // Close the CSV file
    csvFile.close();

    return 0;
}
```

```python
matrix_1 = [0,0,0,0]
for i in range(len(cols)):
    if  cols[i][0]=='C' and dff[cols[i]][0] == 1:
        dff.loc[2][i+1] = "TP"
        matrix_1[0]+=1
    elif cols[i][0]=='N' and dff[cols[i]][0] == 0:
        dff.loc[2][i+1] = "TN"
        matrix_1[1]+=1
    elif cols[i][0]== 'N' and dff[cols[i]][0] == 1:
        dff.loc[2][i+1] = "FP"
        matrix_1[2]+=1
    elif cols[i][0]== 'C' and dff[cols[i]][0] == 0:
        dff.loc[2][i+1] = "FN"
        matrix_1[3]+=1
```

```python
matrix_2 = [0,0,0,0]
for i in range(len(cols)):
    if  cols[i][0]=='C' and dff.loc[1][i+1] == 1:
        dff.loc[3][i+1] = "TP"
        matrix_2[0]+=1
    elif cols[i][0]=='N' and dff[cols[i]][1] == 0:
        dff.loc[3][i+1] = "TN"
        matrix_2[1]+=1
    elif cols[i][0]== 'N' and dff[cols[i]][1] == 1:
        dff.loc[3][i+1] = "FP"
        matrix_2[2]+=1
    elif cols[i][0]== 'C' and dff[cols[i]][1] == 0:
        dff.loc[3][i+1] = "FN"
        matrix_2[3]+=1
```

```python
import matplotlib.pyplot as plt

# The slices will be ordered and plotted counter-clockwise.
# labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
labels = data.iloc[0].index
# sizes = [15, 30, 45, 10]
sizes = data.iloc[0].values
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0, 0)  # explode a slice if required

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=False)
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle
plt.title('Donut chart for the first model')
plt.axis('equal')
plt.show()
```

```python
names_1 = []
for i in range(len(dffff.index)):
    names_1.append(dffff['Unnamed: 0'][i]+'_model')

new_df = pd.DataFrame({
    "TP":[0]*len(names_1),
    "TN":[0]*len(names_1),
    "FP":[0]*len(names_1),
    "FN":[0]*len(names_1), 
    }, index = names_1
)
```

```python
new_df.head(2)
```

```python
for k,j in enumerate(names_1):
    for i in range(len(cols)):
        if  cols[i][0]=='C' and dffff[cols[i]][k] == 1:
            new_df.loc[names_1[k]][0]+=1
        elif cols[i][0]=='N' and dffff[cols[i]][k] == 0:
            new_df.loc[names_1[k]][1]+=1
        elif cols[i][0]== 'N' and dffff[cols[i]][k] == 1:
            new_df.loc[names_1[k]][2]+=1
        elif cols[i][0]== 'C' and dffff[cols[i]][k] == 0:
            new_df.loc[names_1[k]][3]+=1 
```

```python
new_df['TP-FP'] = new_df['TP'] - new_df['FP']
```

```python
myclass = df['class']
myclass
```

```python
dff = df.drop('class', axis=1).transpose()
```

```python
dff.head(2)
```

```python
myfeatures = dff.index
myfeatures
```

```python
dff.columns = myclass
```

```python
mysum = []
for i in range(len(myfeatures)):
    num_ind = dff.loc[myfeatures[i]].sum()
    mysum.append(num_ind)
```

```python
dff['num_ind_per_feature_all'] = mysum
```

```python
not_cancer = ['NC']
everybody = list(dff.columns)
ind_no_can = []
ind_with_can = []
for i in not_cancer:
    for j in everybody[:-1]:
        if (j.find(i)!=-1 and j not in ind_no_can):
            ind_no_can.append(j)
        elif (j.find(i)==-1 and j not in ind_no_can):
            ind_with_can.append(j)
```

```python
print(len(ind_no_can))
print(len(ind_with_can))
```

```python
opq = []
for i in range(len(myfeatures)):
    xyz=0
    for j in range(142):
        xyz += dff[ind_no_can[j]].loc[myfeatures[i]]
    opq.append(xyz)
```

```python
feg = []
for i in range(len(myfeatures)):
    xyz=0
    for j in range(129):
        xyz += dff[ind_with_can[j]].loc[myfeatures[i]]
    feg.append(xyz)
```

```python
len(opq), len(feg)
```

```python
opq[0:20], feg[0:20]
```

```python
max(opq), max(feg)
```

```python
min(opq), min(feg)
```

```python
dff['num_ind_no_can_per_feature'] = opq
```

```python
dff['num_ind_with_can_per_feature'] = dff['num_ind_per_feature_all'] - dff['num_ind_no_can_per_feature']
```

```python
dff['C%'] = dff['num_ind_with_can_per_feature']/len(ind_with_can)
dff['NC%'] = dff['num_ind_no_can_per_feature']/len(ind_no_can)
dff['C% - NC%'] = dff['C%'] - dff['NC%']
dff['C% / NC%'] = dff['C%'] / dff['NC%']
```

```python
dff = dff.rename(columns={"num_ind_per_feature_all": "T", "num_ind_no_can_per_feature": "NC", "num_ind_with_can_per_feature": "C"})
```

```python
dff.to_csv("mutations_v2.csv")
```
